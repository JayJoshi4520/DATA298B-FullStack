```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import os

# Example: Load and display an image
image_path = 'data/cat/cat_image_1.jpg' # Replace with a valid image path from your data
if os.path.exists(image_path):
    try:
        image = Image.open(image_path)
        plt.imshow(image)
        plt.title('Example Image')
        plt.axis('off')
        plt.show()
    except FileNotFoundError:
        print(f"Error: The image file '{image_path}' was not found.")
    except Exception as e:
        print(f"An error occurred while opening the image: {e}")
else:
    print(f"Warning: The path '{image_path}' does not exist. Please create the data directory and a sample image.")
```
